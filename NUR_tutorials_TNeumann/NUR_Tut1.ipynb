```python
### 1.1. exercise: Numerical errors
## a) implementation of sinc(x) in two ways
# power series expansion function (eq 2)
n = 3 # define precission
def sinc(x,n):
    s1 = 0
    frac = 1
    for i in range(1,n+1):
        s1 += (-1)**i*x**(2*i)/frac
        frac = frac * (2*i+1)
        print(frac)
    print(frac)
    return s1
    
# using np.sin(x)
import numpy as np
def sinc_np(x):
    s2 = np.sin(x)/x
    return s2

#comparison via plotting
import matplotlib.pyplot as plt
x = np.linspace(0,10,100) #define x variable
plt.plot(x, sinc_np(x), label = 'numpy sinc(x)')
plt.plot(x, sinc(x,n), label = 'sinc(x)')
plt.xlabel('x')
plt.ylabel('Sinc(x)')
plt.legend()
plt.show()
plt.plot(x, sinc(x,n)/sinc_np(x), label = 'Residual')
plt.xlabel('x')
plt.ylabel('Residuals of sinc(x)/ sinc_np(x)')
plt.legend()
print('the error in this example is a truncation error. It gets smaller with higher values of n. That is reasonable regarding the fluctuations which can be seen in the residuals. means the serie, describing the sinc(x) converge')
plt.show()
```

```python
# 1.1.b) varying n
# 1.1.c) accuracy
n1 = 7
plt.plot(x, sinc_np(x), label = 'numpy sinc(x)')
plt.plot(x, sinc(x,n1), label = 'sinc(x, n=7)')
plt.plot(x, sinc(x,10), label = 'sinc(x, n=10)')
plt.plot(x, sinc(x,2), label = 'sinc(x, n=2)')

plt.xlabel('x')
plt.ylabel('Sinc(x)')
plt.legend()
plt.show()
plt.plot(x, sinc(x,7)/sinc_np(x), label = 'Residual: n=7')
plt.plot(x, sinc(x,10)/sinc_np(x), label = 'Residual: n=10')
plt.plot(x, sinc(x,2)/sinc_np(x), label = 'Residual: n=2')

plt.xlabel('x')
plt.ylabel('Residuals of sinc(x)/ sinc_np(x)')
plt.legend()
print('the error/residuals in this example grow with higher values of n. Through the vary sign of the serial terms, the error is also oscillating.')
plt.show()
```
